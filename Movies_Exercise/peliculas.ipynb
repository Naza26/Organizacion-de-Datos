```python
import pandas as pd

movies = pd.read_csv('./peliculas.csv')
```

```python
# Mostrar si existen datos faltantes (N/A) ¿Qué porcentaje representan del total de datos?
total_data = movies.size
total_null = movies.isnull().sum(numeric_only=True).sum()

porcentaje_nulos = (total_null * 100) / total_data
porcentaje_nulos
```

```python
# Generar un nuevo dataframe dataset_peliculas que contenga las siguientes columnas:
    # original_title
    # title
    # adult
    # original_language
    # popularity
    # budget
    # release_date
    # revenue
    # vote_count
    
new_movies_df = movies[['original_title', 'title', 'adult', 'original_language', 'popularity', 'budget', 'release_date', 'revenue', 'vote_count']].copy()
new_movies_df

# Verificar si existen datos faltantes, en ese caso eliminar las filas que los contengan.
for column in new_movies_df:
    new_movies_df = new_movies_df[new_movies_df[column].notna()]
new_movies_df

# A partir del atributo release_date generar dos nuevos atributos year y month

month = []

year = []

for date in new_movies_df['release_date']:
    date_to_list = date.split('-')
    month.append(date_to_list[1])
    year.append(date_to_list[0])

new_movies_df['year'] = year

new_movies_df['month'] = month

new_movies_df

# Guardar el dataset en un archivo csv con el nombre dataset_peliculas.

df = new_movies_df.to_csv('dataset_peliculas.csv', index=False)

df
```

```python
# ¿Cuál es el idioma más popular de las películas?
# Generar un dataframe idiomas que contenga una columna idioma y otra cantidad donde se pueda ver
# la cantidad de veces que aparece cada idioma en en el dataset.

# Ingles es el idioma mas popular.

# initialize data of lists.
data = {}
  
# Create DataFrame
# df = pd.DataFrame(data)


for l in movies['original_language']:
    if l not in data:
        data[l] = 0
    data[l] = data[l] + 1

data
idioma = {'idioma': [], 'cantidad': []}

for i in data:
    idioma['idioma'].append(i)
    idioma['cantidad'].append(data[i])
idioma
idiomas = pd.DataFrame(idioma, columns=['idioma', 'cantidad'])
idiomas
```

```python
# ¿Qué películas tienen un revenue mayor al promedio? Generar un dataframe mayor_revenue que contenga una columna titulo 
# y otra revenue con las películas que tengan un revenue mayor al promedio.

revenue_promedio = movies['revenue'].mean()

new_df = movies[movies['revenue'] > revenue_promedio]

new_df[['title', 'revenue']]

mayor_revenue = pd.DataFrame(new_df, columns=['title', 'revenue'])
mayor_revenue
```

```python
# ¿Cuál fue la rentabilidad (revenue - budget) de cada película? 
# Generar un dataframe rentabilidad con las columnas titulo, rentabildad, revenue y budget* ordenado de mayor a menor.
# ¿Qué peliculas fueron las más rentables? ¿Cuáles no fueron rentables?

rentabilidad = movies['revenue'] - movies['budget']

data = {'titulo': movies['title'], 'rentabilidad': rentabilidad, 'revenue': movies['revenue'], 'budget': movies['budget']}


new_df = pd.DataFrame(data, columns=['titulo', 'rentabilidad', 'revenue', 'budget'])
new_df.sort_values(by='rentabilidad', ascending=False)
```
